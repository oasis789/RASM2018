```python
from parser import load_image
from plotter import plot_image
from keras import Sequential
from keras.layers import Conv2D, Activation, MaxPooling2D
import numpy as np
from skimage import img_as_float
```

```python
image = img_as_float(load_image('Add MS 7474_0013.tif'))
```

```python
plot_image(image)
```

```python
def plot_convolved(model, image):
    batch = np.expand_dims(image, axis=0)
    conv_image = model.predict(batch)
    conv_image = np.squeeze(conv_image, axis=0)
    print(conv_image.shape)
    conv_image = conv_image.reshape(conv_image.shape[:2])
    print(conv_image.shape)
    plot_image(conv_image)
```

```python
model = Sequential()
model.add(Conv2D(filters=1, kernel_size=(3, 3), input_shape=image.shape))
plot_convolved(model, image)
```

```python
model = Sequential()
model.add(Conv2D(filters=3, kernel_size=(3, 3), input_shape=image.shape))
# model.add(Activation('relu'))
plot_convolved(model, image)
```
